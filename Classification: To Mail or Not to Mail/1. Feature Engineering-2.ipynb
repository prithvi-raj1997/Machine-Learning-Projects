```python
datafile_train=r'D:\Machine Learning\Python Projects\Project-2\carvan_train.csv'
datafile_test=r'D:\Machine Learning\Python Projects\Project-2\carvan_test.csv'

car_train=pd.read_csv(datafile_train)
car_test=pd.read_csv(datafile_test)

print('shape of train data: {}'.format(car_train.shape))
print('shape of test data: {}'.format(car_test.shape))
```

```python
# car_test doen't have outcome variable
# In order to combine the two we need to add response variable to car_test
car_test['V86']=np.nan
```

```python
# creating identifier column for train and test
car_train['data']='train'
car_test['data']='test'

# the columns in two data frames should be in same order to enable the concatination
car_test=car_test[car_train.columns]

car_all=pd.concat([car_train,car_test],axis=0)
car_all.head()
```

```python
car=car_all.copy()

# Returns number of rows and columns in overall data[train and test]
print('shape of overall data: {}'.format(car.shape))
```

```python
car.info()
```

```python
# 'V1'
# there is no order in the sub categories. so label encoding doesn't make sense
# create dummies based on groupby mean on dependent variable

# 'V2', 'V3', 'V4' ---> let it be

# 'V5'
# create dummies based on some frequency cutoff

# L3 : V6 - V43 --> let it be

# L4 : V44 - V64 ---> le it be

# 'V65' to 'V85' ---> le it be
```

```python
# 'V1'
car['V1']=car['V1'].astype(str)

v1=np.round(car.groupby('V1')['V86'].mean(),2).reset_index()

v1=v1.rename({'V86':'V1_V86'},axis=1)

car=pd.merge(car,v1,on='V1',how='left')

car.drop('V1',axis=1,inplace=True)
```

```python
# 'V5'
# create dummies based on some frequency cutoff
car['V5'].value_counts()
```

```python
print('shape of data before creating dummies: {}'.format(car.shape))
```

```python
cat_cols=['V1_V86','V5']

for col in cat_cols:
    freqs=car[col].value_counts()
    k=freqs.index[freqs>50][:-1]
    for cat in k:
        name=col+'_'+str(cat)
        car[name]=(car[col]==cat).astype(int)
        
    del car[col]
    print(col)
```

```python
print('shape of data after creating dummies: {}'.format(car.shape))
```

```python
car.head()
```

```python
# Data preprocessing is complete - the data is in the expected format
```

```python
car_copy=car.copy()
```

Lets separate our two data sets and remove the unnecessay columns that we added while combining them

```python
# Separate the train and test data
car_train=car_copy[car_copy['data']=='train']
car_train.drop('data',axis=1,inplace=True)

car_test=car_copy[car_copy['data']=='test']
car_test.drop(['data','V86'],axis=1,inplace=True)
```

```python
print('shape of train data after feature engineering: {}'.format(car_train.shape))
print('shape of test data after feature engineering: {}'.format(car_test.shape))
```

### Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

#### for train data

```python
feature_scale_train=[feature for feature in car_train.columns if feature not in ['V86']]
len(feature_scale_train)
```

```python
scaler.fit(car_train[feature_scale_train])
```

```python
# transform the train data and add on 'V86' variable

scaled_train=pd.DataFrame(scaler.transform(car_train[feature_scale_train]),
                          columns=feature_scale_train)

data_train=pd.concat([car_train[['V86']].reset_index(drop=True), scaled_train],axis=1)
```

```python
data_train.to_csv('train.csv',index=False)
```

#### for test data

```python
feature_scale_test=[feature for feature in car_test.columns]
len(feature_scale_test)
```

```python
# transform the test data and add on 'V86' variable

scaled_test=pd.DataFrame(scaler.transform(car_test[feature_scale_test]),columns=feature_scale_test)
```

```python
scaled_test.to_csv('test.csv',index=False)
```
