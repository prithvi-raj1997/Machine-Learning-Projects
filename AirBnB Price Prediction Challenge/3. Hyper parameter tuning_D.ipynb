```python
# Hyperparameter tuning for getting best alpha value.run GridSearchCV
ridge=Ridge()
params={'alpha':np.linspace(0.1,5,20)}
ridge_grid=GridSearchCV(ridge,param_grid=params,scoring='neg_root_mean_squared_error',
                       cv=10,verbose=1)
ridge_grid.fit(x_validation,y_validation)
```

```python
def report(results,n_top=3):
    for i in range(1,n_top+1):
        candidates = np.flatnonzero(results['rank_test_score']==i)
        for candidate in candidates:
            print("Model with rank: {0}".format(i))
            print("Mean Validation Score: {0:.8f} (std:{1:.3f})".format(
                results['mean_test_score'][candidate],
                results['std_test_score'][candidate]))
            print("Parameters: {0}".format(results['params'][candidate]))
            print("")
            
```

```python
report(ridge_grid.cv_results_,3)
```

### 5: Gradient Boosting

```python
# Fitting Gradient Boosting on Training set
from hyperopt import fmin, tpe, hp, STATUS_OK, Trials
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import GradientBoostingRegressor
```

#### 5.1: Hyper parameter tuning using Bayesian Hyperopt

```python
# Hyperparameter tuning using Bayesian Hyperopt
def acc_model(params):
    gbm=GradientBoostingRegressor(**params)
    return cross_val_score(gbm,x_validation,y_validation).mean()
```

```python
param_space={
    'learning_rate': hp.choice('learning_rate',[0.01,0.05,0.1,0.4,0.5,1]),
    'n_estimators': hp.choice('n_estimators',range(100,500)),
    'criterion': hp.choice('criterion', ['mse']),
    'max_depth': hp.choice('max_depth', range(1,20)),
    'subsample': hp.choice('subsample',[0.5,0.8,1]),
    'max_features': hp.choice('max_features',range(1,40))
}
```

```python
best=0
def f(params):
    global best
    acc = acc_model(params)
    if acc>best:
        best=acc
    print('new best:',best,params)
    return {'loss':-acc, 'status':STATUS_OK}
```

```python
trials=Trials()
best=fmin(f,param_space,algo=tpe.suggest,max_evals=10,trials=trials)
print('best: {}'.format(best))
```

* 1st best: {'criterion': 0, 'learning_rate': 0, 'max_depth': 8, 'max_features': 17, 'n_estimators': 283, 'subsample': 1}

* 2nd best:{'criterion': 'mse', 'learning_rate': 0.01, 'max_depth': 8, 'max_features': 19,'n_estimators':395, 'subsample': 0.5}

#### 5.2: Fitting the on train data with best parameters and predicing on test data

```python
gbm_bayesian=GradientBoostingRegressor(**{'criterion': 'mse', 'learning_rate': 0.01, 'max_depth': 8, 
                                          'max_features': 19, 'n_estimators':395, 'subsample': 0.5})
gbm_bayesian.fit(x_train,y_train)

predict_gbm=gbm_bayesian.predict(x_test)
print('Average RMSE= {}'.format(np.round(mean_squared_error(y_test,predict_gbm,squared=False), 3)))
print('Explained variance= {}'.format(r2_score(y_test,predict_gbm)))
```

```python
best=0
def f(params):
    global best
    acc = acc_model(params)
    if acc>best:
        best=acc
    print('new best:',best,params)
    return {'loss':-acc, 'status':STATUS_OK}
```
